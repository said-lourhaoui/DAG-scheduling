# Comparison between the two algorithms for differnet DAG parameters

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
results_all= pd.read_csv("results_all.csv")
```

```python
results_all.head(400)
```

# No of nodes influence

```python
nodes_mean = results_all.groupby(["n_nodes"]).mean().reset_index()

plt.plot(nodes_mean["n_nodes"], nodes_mean["GA_makespan"] , label= 'GA' )
plt.plot(nodes_mean["n_nodes"], nodes_mean["proposedAlgo"] , label= 'proposed algo' )
plt.xlabel("No of nodes")
plt.ylabel("Makespan")
plt.title("Makespan vs No of nodes for both algorithms")
plt.legend()
plt.show()
#nodes_mean.head()
```

```python
nodes_mean.plot(x="n_nodes", y = ["GA_makespan", "proposedAlgo"], kind='bar', label = ["GA", 'proposedAlgo'] )
plt.xlabel("No of nodes")
plt.ylabel("Makespan")
plt.title("Makespan vs No of nodes for both algorithms")
plt.show()
```
