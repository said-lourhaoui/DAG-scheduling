# Comparison between different parameters of the genetic algorithm

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Ga_results = pd.read_csv("results_GA.csv")
```

```python
Ga_results.head()
```

```python
chromosomes_mean =  Ga_results.groupby(['n_chromosomes']).mean()
plt.plot(chromosomes_mean.index, chromosomes_mean['GA_makespan'])
plt.xlabel("no of chromosomes ") 
plt.ylabel("makespan")
plt.title("Makespan vs number of chromosomes")
plt.show()
#chromosomes_mean.head()
```

```python
gen_mean =  Ga_results.groupby(['best_generation']).mean()
plt.plot(gen_mean.index, gen_mean['GA_makespan'])
plt.xlabel("no of generations") 
plt.ylabel("makespan")
plt.title("Makespan vs number of generations ")
plt.show()
#gen_mean.head()
```

```python
X_mean =  Ga_results.groupby(['X_rate']).mean()
plt.plot(X_mean.index, X_mean['GA_makespan'])
plt.xlabel("Crossover rate") 
plt.ylabel("makespan")
plt.title("Makespan vs crossover rate")
plt.show()
#X_mean.head()
```

```python
M_mean =  Ga_results.groupby(['M_rate']).mean()
plt.plot(M_mean.index, M_mean['GA_makespan'])
plt.xlabel("Mutation rate") 
plt.ylabel("makespan")
plt.title("Makespan vs mutation rate")
plt.show()
#M_mean.head()
```
